```python
# 1.2 张量的维度和形变
```

```python
# 创建高位张量 查看维度和形状

torch.tensor([1.2, 2.3]).ndim ,  torch.tensor((1.2, 2.3)).shape
```

```python
# 返回有几个n-1维的元素
len(torch.tensor([1.1, 2.2, 3.3]))

# 返回总共有几个数字
torch.tensor([1.2,2.3,3.4,4.5]).numel()

# list的list创建
t2=torch.tensor([[1,2],[3,4]])
t2, t2.ndim, t2.shape, len(t2), t2.numel()
```

```python
# 高维张量
a1=np.array([[1,2,2],[3,4,4]]) 
a1
a2=np.array([[5,6,6],[7,8,8]])
a2

# 由两个形状相同的3维数组创建一个3维的张量
t3=torch.tensor([a1, a2])
t3
# t3.ndim, t3.shape, len(t3), t3.numel()
```

```python
# flatten 拉平：将任意维度张量转化为一维张量
t3.flatten()

# reshape:任意变形
t3.reshape(12,)
```

```python
# 1.3 特殊张量的创建方法
```

```python
# 全0张量
torch.zeros([3,2])

# 全1张量
torch.ones([2,3])

# 单位阵
torch.eye(5)

# 对角矩阵  不能使用list直接创建对角矩阵，必须使用tensor
torch.diag(torch.tensor([2,4,6,8]))

# rand:服从均分分布的张量
torch.rand([3,2])

# randn:服从标准正态分布的张量
torch.randn([3,4])

# normal:服从正态分布的张量 均值为2，std为3的normal
torch.normal(2,3,size=(2,3,4))

# randint:整数随机采样
torch.randint(0,10,size=(2,4,5))

# arange/linspace生成数列
torch.arange(1,5), torch.arange(1,5,0.5), torch.linspace(1,5,4)

# empty:生成未初始化的指定形状矩阵
torch.empty([3,4])

# full:根据指定形状，填充指定数值
torch.full([3,4],521)
```

```python
# 创建指定形状的数组 _like前后张量数据类型应一致
torch.full_like(torch.tensor([[[2,2,3],[1,2,4],[5,4,7]]]) , 521 ), torch.randint_like(torch.tensor([[1,2],[2,3]]),5,10), 
torch.zeros_like(torch.tensor([1,2])), torch.randn_like(torch.tensor([2.0, 5.0]))
```

```python
# 1.4 张量和其他相关类型之间的转化方法
```

```python
# numpy:张量转数组    tensor to array
t1=torch.tensor([[1,2],[3,4]]) 
t1.numpy(), np.array(t1)

# tolist:张量转列表 tensor to list  注意区分tolist()和list()
t1.tolist()
list(t1)

# item:转化为数值 只能转化one element
torch.tensor([2]).item()
```

```python
# 1.5 张量的深拷贝
```

```python
# 如果只用等号则是浅拷贝，改变一个另一个也会变；使用clone则是深拷贝，改变一个另一个不会变，两者相互独立
t10=t1.clone()
t10[0]
```
